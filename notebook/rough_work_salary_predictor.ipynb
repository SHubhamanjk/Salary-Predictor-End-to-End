```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import BayesianRidge


models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    ElasticNet(),
    SVR(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    AdaBoostRegressor(),
    ExtraTreesRegressor(),
    XGBRegressor(),
    LGBMRegressor(),
    KNeighborsRegressor(),
    BayesianRidge(),
]

scores=[]
for model in models:
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    score = model.score(X_test, y_test)
    scores.append(score)
    print(f"The score of the {model.__class__.__name__} on the test data is: {score}")
```

```python
import numpy as np

def process_user_input(X_test_columns):
    age = int(input("Enter age: "))
    education_level_input = input("Enter education level (Bachelors, Masters, PhD): ").strip().lower()
    years_of_experience = int(input("Enter years of experience: "))
    job_title_input = input("Enter job title: ").strip().lower().replace(" ", "")

    # Map the education level to numerical values
    if education_level_input in ["bachelors", "bachelor's"]:
        education_level = 0
    elif education_level_input in ["masters", "master's"]:
        education_level = 1
    elif education_level_input == "phd":
        education_level = 2
    else:
        raise ValueError("Invalid education level. Please enter 'Bachelors', 'Masters', or 'PhD'.")

    # Initialize an array with zeros
    input_data = np.zeros(len(X_test_columns))

    # Set the basic input data
    input_data[0] = age
    input_data[1] = education_level
    input_data[2] = years_of_experience

    # Handle the job title input
    job_title_col = f"jobtitle_{job_title_input}"
    if job_title_col in X_test_columns:
        job_title_index = X_test_columns.get_loc(job_title_col)
        input_data[job_title_index] = 1
    else:
        raise ValueError(f"Job title '{job_title_input}' not found in the training data.")

    return input_data

def salary_predictor(model, X_test_columns):
    try:
        input_data = process_user_input(X_test_columns)
        expected_salary = model.predict([input_data])[0]
        print(f"Expected Salary is {expected_salary}")
        print()
        print()
    except ValueError as e:
        print(e)
```

```python
salary_predictor(model, X_test.columns)
```
