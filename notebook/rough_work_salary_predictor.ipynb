```python
import numpy as np

def process_user_input(X_test_columns):
    age = int(input("Enter age: "))
    education_level_input = input("Enter education level (Bachelors, Masters, PhD): ").strip().lower()
    years_of_experience = int(input("Enter years of experience: "))
    job_title_input = input("Enter job title: ").strip().lower().replace(" ", "")

    # Map the education level to numerical values
    if education_level_input in ["bachelors", "bachelor's"]:
        education_level = 0
    elif education_level_input in ["masters", "master's"]:
        education_level = 1
    elif education_level_input == "phd":
        education_level = 2
    else:
        raise ValueError("Invalid education level. Please enter 'Bachelors', 'Masters', or 'PhD'.")

    # Initialize an array with zeros
    input_data = np.zeros(len(X_test_columns))

    # Set the basic input data
    input_data[0] = age
    input_data[1] = education_level
    input_data[2] = years_of_experience

    # Handle the job title input
    job_title_col = f"jobtitle_{job_title_input}"
    if job_title_col in X_test_columns:
        job_title_index = X_test_columns.get_loc(job_title_col)
        input_data[job_title_index] = 1
    else:
        raise ValueError(f"Job title '{job_title_input}' not found in the training data.")

    return input_data

def salary_predictor(model, X_test_columns):
    try:
        input_data = process_user_input(X_test_columns)
        expected_salary = model.predict([input_data])[0]
        print(f"Expected Salary is {expected_salary}")
        print()
        print()
    except ValueError as e:
        print(e)
```

```python
salary_predictor(model, X_test.columns)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/content/Dataset09-Employee-salary-prediction.csv")
data.dropna(inplace=True)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Education Level'] = le.fit_transform(data['Education Level'])
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)
job_title_encoded = one_hot_encoder.fit_transform(data[['Job Title']])
df_one_hot = pd.DataFrame(job_title_encoded, columns=one_hot_encoder.get_feature_names_out(['Job Title']))
data = pd.concat([data, df_one_hot], axis=1)
data.drop(['Job Title'],axis=1,inplace=True)
data.drop(['Gender'],axis=1,inplace=True)
if data.isnull().any().any()==True:
  data.dropna(inplace=True)
data.columns = data.columns.str.replace(' ', '', regex=False).str.lower()
X= data.drop(['salary'],axis=1)
y=data['salary']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
from sklearn.linear_model import Ridge
model=Ridge()
model.fit(X_train,y_train)
model.score(X_test, y_test)
data.drop(['salary'],axis=1,inplace=True)
import numpy as np

def process_user_input(X_test_columns):
    age = int(input("Enter age: "))
    education_level_input = input("Enter education level (Bachelors, Masters, PhD): ").strip().lower()
    years_of_experience = int(input("Enter years of experience: "))
    job_title_input = input("Enter job title: ").strip().lower().replace(" ", "")

    # Map the education level to numerical values
    if education_level_input in ["bachelors", "bachelor's"]:
        education_level = 0
    elif education_level_input in ["masters", "master's"]:
        education_level = 1
    elif education_level_input == "phd":
        education_level = 2
    else:
        raise ValueError("Invalid education level. Please enter 'Bachelors', 'Masters', or 'PhD'.")

    # Initialize an array with zeros
    input_data = np.zeros(len(X_test_columns))

    # Set the basic input data
    input_data[0] = age
    input_data[1] = education_level
    input_data[2] = years_of_experience

    # Handle the job title input
    job_title_col = f"jobtitle_{job_title_input}"
    if job_title_col in X_test_columns:
        job_title_index = X_test_columns.get_loc(job_title_col)
        input_data[job_title_index] = 1
    else:
        raise ValueError(f"Job title '{job_title_input}' not found in the training data.")

    return input_data

def salary_predictor(model, X_test_columns):
    try:
        input_data = process_user_input(X_test_columns)
        expected_salary = model.predict([input_data])[0]
        print(f"Expected Salary is {expected_salary}")
        print()
        print()
    except ValueError as e:
        print(e)
```
